### 1. [Importing Dependencies](#importing-dependencies)
In this section, we will import all the necessary Python packages to execute the code cells below.

```python
from datetime import datetime, timedelta
import pandas as pd
from openbb import obb

# Setting output preferences for OpenBB
obb.user.preferences.output_type = "dataframe"

print("Dependencies imported successfully.")
```

### 2. Downloading Historical Stock Prices

We will download historical prices for the SPY ETF using OpenBB with the "yfinance" provider.

```python
# Downloading historical stock prices for SPY
spy_data = obb.equity.price.historical(symbol="SPY", provider="yfinance")

# Display the first few rows of the data
print(spy_data.head())

# Display a summary of the data
print(spy_data.describe())
```

### 3. Analyzing Balance Sheet Metrics

Next, we'll gather balance sheet metrics for Apple Inc. (AAPL) and Microsoft Corp (MSFT) using OpenBB and "yfinance" as the provider.

```python
# Fetching balance sheet metrics for Apple and Microsoft
aapl_msft_metrics = obb.equity.fundamental.metrics(
    "AAPL,MSFT",
    provider="yfinance")

# Displaying the metrics
print(aapl_msft_metrics)
```

### 4. Screening Stock Groups

Using OpenBB, we will now analyze different stock groups using various metrics such as valuation and performance from the "finviz" provider.

```python
# Overview screener: Valuation metrics for industries
industry_valuation = obb.equity.compare.groups(
    group="industry",
    metric="valuation",
    provider="finviz")

print(industry_valuation)
```

```python
# Overview screener: Performance metrics for technology group
technology_performance = obb.equity.compare.groups(
    group="technology",
    metric="performance",
    provider="finviz")

print(technology_performance)
```

```python
# Overview screener: Overview metrics for sectors
sector_overview = obb.equity.compare.groups(
    group="sector",
    metric="overview",
    provider="finviz")

print(sector_overview)
```
